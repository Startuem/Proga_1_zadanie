### Часть 3 

Затем вы можете преобразовать значения температуры из Фаренгейта в Цельсии.

- Создайте новый столбец в `selected` с именем `Celsius`
- Преобразуйте температуру в Фаренгейтах из `TEMP` используя формулу преобразования ниже и сохраните результаты в новой колонке `Celsius`.

$$
\Large
\begin{equation}
  T_{\mathrm{Celsius}} = (T_{\mathrm{Fahrenheit}} - 32)~/~1.8
\end{equation}
$$

- Округлите значение в колонке `Celsius` до 0 знаков после запятой (**не** создавайте новую колонку, обновите текущую)
- Преобразуйте значения `Celsius` в целочисленные (**не** создавайте новую колонку, обновите текущую)

```python
selected.round(2) # Это нужно для задания в конце - округлить все данные до 2 десятичных значения для данных с плавающей запятой
```

```python
selected['Celsius'] = 0.0 # создал столбец Celsius
for idx, row in selected.iterrows():
    TEMP_C = (row['TEMP'] - 32)/1.8 # для каждой строчки в столбце TEMP перевел данные из фаренгейта в цельсий
    selected.at[idx, 'Celsius'] = TEMP_C # добавил значения цельсия в пустой пока что столбец Celsius
selected.Celsius = selected.Celsius.round() # округлил зачения в Celsius
selected.Celsius = selected.Celsius.astype(int) # преобразовал значения в целочисленные
```
